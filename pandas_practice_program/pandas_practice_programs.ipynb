```python
# 1 create and display a one-dimensional array-like object containing an array of data using Pandas module.
import pandas as pd
import numpy as np
arr=[1,2,3,4,5]

series=pd.Series(arr)
print(series)
```

```python
# 2 to convert a Panda module Series to Python list and it's type.

print(series)
print("Series to List")
arr=series.to_list()
print(arr)
print(type(arr))
```

```python
#3.to add, subtract, multiple and divide two Pandas Series.

series1 = pd.Series([2, 4, 6, 8, 10])
series2 = pd.Series( [1, 3, 5, 7, 9])
print("Addition")
print(series1 + series2)
print("\nSubtraction")
print(series1-series2)
print("\nMultiplication")
print(series1 * series2)
print("\nDivision")
print(series1/series2)
```

```python
 # 4 to get the powers of an array values element-wise.
a=pd.Series([0,1,2,3,4,5,6])
b=pd.Series([3,3,3,3,3,3,3])

print("Original Array")
print(a.values)
print(b.values)

print("First array elements raised to powers from second array, element-wise:")
print((a**b).values)
```

```python
#5.create and display a DataFrame from a specified dictionarydata which has the index labels.
exam_data = {'name': ['Anastasia','Dima','Katherine','James','Emily','Michael','Matthew','Laura','Kevin','Jonas'],
             'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
             'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
             'qualify': ['yes','no','yes','no','no','yes','yes','no','no','yes']}


labels=['a','b','c','d','e','f','g','h','i','j']

df=pd.DataFrame(exam_data,labels)
print(df)
```

```python
#6 to display a summary of the basic information about a specified Data Frame and its data.

print(df.info())
print(df.describe())
```

```python
# 10 to select the rows where the number of attempts in the examination is greater than 2.

df2=df["attempts"]>2
print(df[df2])
```

```python
# 11 to count the number of rows and columns of a DataFrame.

print("Number of rows : ",end=" ")
rows,columns=df.shape
print(rows)
print(f"Number of columns : {columns}")
```

```python
# 12.to select the rows where the score is missing, i.e. is NaN.

print(df[df.isnull().any(axis=1)] )
```

```python
#13. to select the rows where number of attempts in the examination is less than 2 and score greater than 15.

df2=(df["attempts"]< 2) & (df["score"]> 15)
print(df[df2])
```

```python
#15. to calculate the sum of the examination attempts by the students.

print(df["attempts"].sum())
```

```python
#16 to calculate the mean score for each different student inDataFrame.

mean_value=df["score"].mean()
print(mean_value)
```

```python
# 17. to append a new row 'k' to data frame with given values for each column.Now delete the new row and return the original DataFrame.

df.loc["k"]={"name":"Suresh", "score": 15.5, "attempts": 1, "qualify": "yes", "label": "k"}
print("Added a row k")
print(df)
print("\nDeleted the row k")
df=df.drop("k")
print(df)
```

```python
#18.to sort the DataFrame first by 'name' in descending order, then by 'score' in ascending order.

print("sort by name")
print(df.sort_values("name",ascending=False))

print("\nsort by score")
print(df.sort_values("score"))
```

```python
# 19.to replace the 'qualify' column contains the values 'yes' and'no' with True and False.

df["qualify"]=df["qualify"].replace({"no":False,"yes":True})
print(df)
```

```python
# 20.to delete the 'attempts' column from the DataFrame.

df=df.drop(columns="attempts")
print(df)
```
