```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# 1. Write a program to draw a scatter plot for random 1000 x and y coordinates

np.random.seed(1)
x=np.random.rand(1000)
y=np.random.rand(1000)

df = pd.DataFrame({"X": x, "Y": y})
fig=px.scatter(df,x="X",y="Y",title="Scatter Plot")
fig.show()
```

```python
#2. Write a program to draw line and scatter plots for random 100 x and y coordinates

np.random.seed(2)
x = np.arange(100)
y = np.random.randn(100)
df = pd.DataFrame({"X": x, "Y": y})
fig=px.scatter(df,x="X",y="Y",title="Line and Scatter Plot")
fig.update_traces(mode="lines+markers")
fig.show()
```

```python
# 3. Write a program to draw a scatter plot for random 500 x and y coordinates and style it
x = np.random.rand(500)
y = np.random.rand(500)

df = pd.DataFrame({
    "X": x,
    "Y": y,
})

fig = px.scatter(
    df,
    x="X",
    y="Y",
    color= x + y,      
    size= np.random.rand(500),         
    title="Styled Scatter Plot ",
   
)

fig.update_layout(
    xaxis_title="X Values",
    yaxis_title="Y Values"
)

fig.show()
```

```python

import plotly.graph_objects as go
df = pd.read_csv("/Users/abinkgeo/BridgeLabz/code/plotly_practice_program/usa_states.csv")

fig = go.Figure(
    data=go.Scatter(
        x = df["Postal"],
        y = df["Population"],
        mode = "markers",
        marker=dict(
            size=12,
            color=df["Population"], 
            colorscale="Viridis",
            showscale=True
        ),
        text = df["State"],           
        hovertemplate="<b>%{text}</b><br>Population: %{y}<extra></extra>"
    )
)

fig.update_layout(
    title="USA States Population (2014)",
    xaxis_title="State Code",
    yaxis_title="Population"
)

fig.show()
```
