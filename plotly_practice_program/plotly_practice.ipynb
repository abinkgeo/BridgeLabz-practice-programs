```python
import plotly.express as px
import pandas as pd

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

fig = px.line(x=x, y=y, title=" Line Chart")
fig.show()
```

```python
#Line Chart
df = pd.DataFrame({
    "Year": [2019, 2020, 2021, 2022],
    "Sales": [200, 250, 300, 400]
})


fig=px.line(df,x="Year",y="Sales",title="Sales over the years")
fig.show()
```

```python
#Bar chart
df = pd.DataFrame({
    "Product": ["A", "B", "C"],
    "Revenue": [500, 300, 700]
})


fig=px.bar(df,x="Product",y="Revenue",color="Product")
fig.show()
```

```python
#Scatter plot

df = pd.DataFrame({
    "Hours": [1, 2, 3, 4, 5],
    "Marks": [40, 50, 65, 75, 90]
})


fig=px.scatter(df,x="Hours",y="Marks",title="Study hours vs marks",)
fig.show()
```

```python
#pie chart

df = pd.DataFrame({
    "Type": ["Organic", "Sponsored"],
    "Count": [70, 30]
})


fig=px.pie(df,names="Type",values="Count",title="Product Distribution",hole=0.4)
fig.show()
```

```python
#box plot
df = pd.DataFrame({
    "Department": ["IT", "IT", "HR", "HR"],
    "Salary": [60000, 65000, 40000, 42000]
})

fig = px.box(
    df,
    x="Department",
    y="Salary",
    title="Salary Spread"
)
fig.show()
```
