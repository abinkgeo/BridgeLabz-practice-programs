```python
#1. to create a pie chart of the popularity of programming
import matplotlib.pyplot as plt
import pandas as pd

languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.pie(popularity, labels=languages, autopct='%1.1f%%')
plt.show()
```

```python
#2.to create a pie chart with a title of the popularity of programming
plt.pie(popularity, labels=languages, autopct='%1.1f%%', startangle=140)
plt.title('Popularity of Programming Languages')
plt.show()
```
